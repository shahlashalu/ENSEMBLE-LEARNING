```python
df = pd.read_csv('/content/drive/MyDrive/DATA SCIENCE/MACHINE LEARNING/MACHINE LEARNING ALGORITHMS/ENSEMBLE CONCEPT/Boosting/AdaBoosting/Iris.csv')
```

```python
x = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
x.head()
```

```python
y = df['Species']
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Import train_test_split function

from sklearn.model_selection import train_test_split


# Split dataset inte training set and test set

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3)


# Import the Adamoest classifier

from sklearn.ensemble import AdaBoostClassifier


# Create AdaBoostClassifer object

abc = AdaBoostClassifier (n_estimators = 50, learning_rate = 1, random_state = 0)          # learning rate is a hyperparameter that controls the contribution of each weak learner to the final combination


# Train AdaBoost classifier

model1 = abc.fit(x_train,y_train)


# Predict the response for test dataset

y_pred = model1.predict(x_test)
```

```python
# import scilit -learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score


# calculate and print model accuracy
print("AdaBoost Classifier Model Accuracy : ",accuracy_score(y_test,y_pred))


```

**Furthur Evaluation with SVC Base Estimator**

```python
#load required classifier
from sklearn.ensemble import AdaBoostClassifier

#import support vector classifier
from sklearn.svm import SVC

#import scikit-learn metrics module for accuracy score
from sklearn.metrics import accuracy_score
svc = SVC(probability = True, kernel = 'linear')

#create adaboost classifier object
abc = AdaBoostClassifier(n_estimators=50, estimator=svc, learning_rate=1, random_state=0)   # Base Estimator : By setting base_estimator = svc, you specify that the base estimator for AdaBoost Classifier is svc

#train AdaBoost Classifier
model2 = abc.fit(x_train,y_train)

#predict the response for test dataset
y_pred = model2.predict(x_test)
```
